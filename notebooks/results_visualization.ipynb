```python
import pandas as pd
import matplotlib.pyplot as plt
import cv2
import os

# Load training metrics
metrics = pd.read_csv("../results/metrics.csv")
```

```python
# Loss Plot
plt.plot(metrics["epoch"], metrics["train_loss"], label="Train Loss")
plt.plot(metrics["epoch"], metrics["val_loss"], label="Validation Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Training vs Validation Loss")
plt.legend()
plt.grid(True)
plt.show()

# PSNR & SSIM Plot
plt.plot(metrics["epoch"], metrics["psnr"], label="PSNR")
plt.plot(metrics["epoch"], metrics["ssim"], label="SSIM")
plt.xlabel("Epoch")
plt.ylabel("Score")
plt.title("PSNR & SSIM Over Epochs")
plt.legend()
plt.grid(True)
plt.show()
```
